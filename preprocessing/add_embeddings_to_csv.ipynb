## To add extracted image embeddings to the csv text dataset

```python
import os
import numpy as np
import pandas as pd
from tqdm import tqdm
folder_path = "C:/Users/gargi/Downloads/68e8d1d70b66d_student_resource/embeddingsfinal"

data = []
```

```python
for file in tqdm(os.listdir(folder_path)):
    if file.endswith(".npy"):
        full_path = os.path.join(folder_path, file)
        arr = np.load(full_path)

        # Replace suffix and add .jpg
        base_name = file.replace("_embedding.npy", ".jpg")

        arr = arr.flatten()

        data.append({
            "filename": base_name,
            **{f"emb_{i}": arr[i] for i in range(len(arr))}
        })

df_embeddings = pd.DataFrame(data)
```
